```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv("Wholesale customers data.csv")
print("first five rows:")
df.head(5)

```

```python
print("last five rows:")
df.tail(5)
```

```python
print("column info:")
df.info()
```

```python
print("shape of dataset:", df.shape)
print(df.columns)
print(df.keys())
```

```python
print("summary stats:")
print(df.describe())
```

```python
pd.unique(df['Channel'])  # Capital C

```

```python
df= df.drop(["Channel","Region"], axis=1) # dropping irrelevent columns
df.head()


```

```python
df = df.rename(columns={
    'Fresh': 'fresh',
    'Milk': 'milk',
    'Grocery': 'grocery',
    'Frozen': 'frozen',
    'Detergents_Paper': 'detergents_paper',
    'Delicassen': 'delicassen',
    'Channel': 'channel',
    'Region': 'region'
})

print(df.columns)
```

```python
# dropping duplicate rows
print(df.shape)
duplicate_rows= df[df.duplicated()]
print("number of duplicate rows:", len(duplicate_rows))
print(df.head(2))
print("after deleting duplicate rows the shape of df is:",df.shape)

(1599, 10)
```

```python
df = df.drop_duplicates()
df.head()
```

```python
df.count()
```

```python
# dropping the missing or null values
print(df.isnull().sum())

```

```python
df = df.dropna()    # Dropping the missing values.
df.count()
```

```python
print(df.isnull().sum())   # After dropping the values

```

```python
sns.boxplot(x=df['milk'])
```

```python
sns.histplot(df['grocery'], kde=True)  # kde=True adds a smooth curve
plt.title("Distribution of Grocery Spend")
plt.xlabel("Grocery Spend")
plt.ylabel("Count")
```

```python
sns.scatterplot(x='milk', y='grocery', data=df)
plt.title("Scatter Plot: Milk vs Grocery Spend")
plt.xlabel("Milk")
plt.ylabel("Grocery")
```

```python
sns.lineplot(x='frozen', y='delicassen', data=df)
plt.title("Line Plot: Frozen vs Delicassen Spend")
plt.xlabel("Frozen")
plt.ylabel("Delicassen")
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Feature Correlation Heatmap")
plt.show()
```
